```python
from openai import OpenAI
client = OpenAI()
```

```python
# Create character list and analysis function
def create_character_list(analysis):
    prompt = f"Create a bulleted list of characters for the following book and provide a brief 2 sentence synopsis of each character: {(analysis)}"

    return prompt
```

```python
# Function Test
analysis = create_character_list("The Lord of the Rings by J.R.R. Tolkien")

print(analysis)
```

```python
# Reach out to OpenAI for the answer

response = client.chat.completions.create(
    model = "gpt-3.5-turbo",
    messages = [
        {
            "role": "system",
            "content": analysis
        },
    ],
    temperature = 0.8,
    top_p = 1,
)

print(response.choices[0].message.content)
```

```python
# Putting the characters into a list
import re

text = response.choices[0].message.content

pattern = re.compile(r'- (.+)')
matches = pattern.findall(text)

archetype_list = []

for match in matches:
    archetype_list.append(match)

print(archetype_list)
```

```python
# Sending the item to DALLE 
image = client.images.generate(
    model ="dall-e-3",
    prompt=archetype_list[0],
    size="1024x1024",
    quality="standard",
    n=1,
)

image_url = image.data[0].url 

print(image.data[0].url)
```
