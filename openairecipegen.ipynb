```python
import os

# Read API key from file and set it for this notebook session
with open("DALLE_Key") as f:
    os.environ["OPENAI_API_KEY"] = f.read().strip()
```

```python
from openai import OpenAI
client = OpenAI()
```

```python
#create shopping list and recipe functions

def create_shopping_list(recipe):
    prompt = f"create a bulleted shopping list for the following recipe and a wine or beer that pairs well with this particular meal and cooking instructions: {(recipe)}"

    return prompt
```

```python
# function test

recipe = create_shopping_list("fettuccine alfredo with chicken and broccoli")

print(recipe)
```

```python
# reaching out to openai for answer

response = client.chat.completions.create(
    model = "gpt-3.5-turbo",
    messages = [
        {
            "role": "system",
            "content": recipe
        },
    ],
    temperature = 0.7,
    top_p = 1,
)

print(response.choices[0].message.content)
```

```python
import re

text = response.choices[0].message.content

pattern = re.compile(r'- (.+)')
matches = pattern.findall(text)

shopping_list = []

for match in matches:
    shopping_list.append(match)

print(shopping_list)
```

```python
# generate images using DALL_E
image = client.images.generate(
    model ="dall-e-3",
    prompt =shopping_list[0],
    size ="1024x1024",
    quality ="standard",
    n=1,
)

image_url = image.data[0].url

print(image.data[0].url)
```
